```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:/Users/carla/Desktop/SSMS_Projects/resources/Customer Call List.xlsx')
df
```

```python
#IMPORTANT: Before assigning 'df =' in front of any changes, review the output first to avoid saving undesired alterations
```

```python
#Delete redundant column
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
#Remove duplicates
df = df.drop_duplicates()
df
```

```python
#Strip to remove undesired symbols on 'Last_Name' column
#Note: When cleaning the 'Last_Name' column using df['Last_Name'].str.strip("..._/ "), remember to assign the cleaned values only to that column using df['Last_Name'] =. Failing to do so might inadvertently apply these changes to the entire table, affecting other columns unintentionally
df.loc[:, 'Last_Name'] = df['Last_Name'].str.strip("..._/ ")
df
```

```python
#Standarize 'Phone_Number'
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df
```

```python
#Split Address info
#Split address into 3 different columns
address_split = df['Address'].str.split(',', n=2, expand=True)

#rename columns
address_split.columns = ['Street_Address', 'City', 'Postal_Code']

#Using .loc to asign new values to the columns
df.loc[:, ['Street_Address', 'City', 'Postal_Code']] = address_split
df
```

```python
#Standarizing 'Paying Customer' column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No')
```

```python
df = df.replace('No/a', '')
df = df.fillna('')
```

```python
#Standarizing 'Do_Not_Contact'
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', 'No')
```

```python
df
```

```python
#We now want to drop all the rows from people who DO NOT want to be contacted
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace = True)
df
```

```python
#Do the same thing for blank Phone_Numbers
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
df
```

```python
#Reorganize index
df.reset_index()
df = df.reset_index(drop=True)
```
